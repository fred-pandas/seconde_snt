```python
import osmnx as ox
import folium
```

```python
départ = [48.9611488,2.8745053]
```

```python
carte = folium.Map(location=départ,tiles='OpenStreetMap', zoom_start=15)
carte
```

```python
folium.Marker(départ,
              popup='départ',
             icon=folium.Icon(color='blue')).add_to(carte)
```

```python
point = ox.geocode('41 A cours pinteville, 77100 Meaux')
point
```

```python
G = ox.graph_from_address('Meaux, France',distance=1000, network_type='drive')
```

```python
ox.plot_graph_folium(G,edge_color='red',edge_width=1,graph_map=carte,popup_attribute='length')
```
